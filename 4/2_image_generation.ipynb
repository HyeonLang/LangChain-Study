#이미지에서 텍스트 편집

```python
#이미지와 마스크 준비 (현재 폴더에 이미지와 마스크를 넣어 둡니다.)
image = open("image.png", "rb")
mask = open("mask.png", "rb")

#프롬프트 준비
prompt = "many apples in cardboard box"
```

```python
import openai

#이미지와 텍스트에서 이미지 편집을 실행
response = openai.Image.create_edit(
    image=image,
    mask=mask,
    prompt=prompt,
    n=1,
    size="512x512"
)
image_url = response["data"][0]["url"]
print(image_url) #이미지 사용기한 1시간
```

```python
print(response)
```

# 이미지에서 변형 생성

```python
#이미지와 준비 (현재 폴더에 이미지를 넣어 둡니다.)
image = open("image.png", "rb")
```

```python
import openai

#이미지와 텍스트에서 이미지 편집을 실행
response = openai.Image.create_variation(
    image=image,
    n=1,
    size="512x512"
)
image_url = response["data"][0]["url"]
print(image_url) #이미지 사용기한 1시간
```

```python
print(response)
```
